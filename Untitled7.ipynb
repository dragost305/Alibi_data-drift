```python
# Import libraries
import numpy as np
from sklearn.cluster import KMeans
from alibi_detect.cd import KSDrift
import matplotlib.pyplot as plt
```

```python
# Generate initial data
np.random.seed(0)
data_dim = 5
n_samples = 1000
X_train = np.random.rand(n_samples, data_dim)
```

```python
import numpy as np

# Set seed for reproducibility
np.random.seed(1)

# Define the percentage of samples to be affected by drift
drift_percentage = 0.2

# Create a copy of the initial dataset
X_drift = X_train.copy()

# Randomly select indices for the samples to be affected by drift
drifted_indices = np.random.choice(len(X_train), size=int(drift_percentage * len(X_train)), replace=False)

# Add outliers to the selected samples
X_drift[drifted_indices] = np.random.rand(len(drifted_indices), X_train.shape[1]) + 2.0
```

```python
# Train KMeans model
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X_train)
```

```python
# Initialize drift detector
# p_val determines the threshold for considering a change in the data distribution as statistically significant.
detector = KSDrift(X_train, p_val=.05)
```

```python
# Detect drift - checks wheather the samples have drift or not

# Tak a batch of samples as input and compares their distribution to the reference (initial) distribution.i.e. X_drift
preds_drift = detector.predict(X_drift)
```

```python
# Visualize the data
plt.scatter(X_train[:, 0], X_train[:, 1], label='Initial Data', alpha=0.7)
plt.scatter(X_drift[:, 0], X_drift[:, 1], label='Data with Drift', alpha=0.7)
plt.title('Data Drift Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
